- Target variable is highly imbalanced. Approximately 91.93% of applicants belong to class 0 (non-default) and only 8.07% belong to class 1 (default). - This imbalance means that accuracy alone is not a reliable evaluation metric. Therefore, stratified train-test splitting should be used to preserve     class proportions, and evaluation metrics such as ROC-AUC, Precision, Recall, F1-Score, and Precision-Recall AUC should be considered.

##### Missing Value Analysis

```python
missing=df.isna().sum()
missing = missing[missing > 0]
missing.head(20)
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0]

missing_df = pd.DataFrame({
    'Missing_Count': missing,
    'Missing_Percentage': missing/len(df)*100
})

missing_df.sort_values(
    by='Missing_Percentage',
    ascending=False
).head(20).plot(
    kind='barh',
    figsize=(10,8)
)
plt.show()
```
